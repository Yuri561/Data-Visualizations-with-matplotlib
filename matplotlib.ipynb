```python
# this is how you change your linestyle on matplot 

import numpy as np 
import matplotlib.pyplot as plt 
x = np.linspace(0,10,1000)
plt.plot(x,x + 0,linestyle = 'solid')
plt.plot(x,x +1, linestyle = 'dashed')
plt.plot(x, x+2, linestyle = 'dashdot')
plt.plot(x,x + 3, linestyle = 'dotted')
```

```python

#this is another way you can specify the colors to your graph 

plt.plot(x,x + 0,'-g')
plt.plot(x, x +1,'--c')
plt.plot(x,x+2, '-.k')
plt.plot(x,x + 3,':r')
```

```python
#how to change the values of the x and y axis 

plt.plot(x, np.sin(x))
plt.xlim(-1,11)
plt.ylim(-1.5, 1.5)
```

```python
plt.plot(x, np.sin(x))
plt.xlim()
plt.ylim(0,100)
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal');
```

```python
#you can use ; to close your arguement this is a way to write codes professionally


plt.plot(x, np.sin(x))
plt.title('A Sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)');
```

```python
#this is how you create legends using cos sine and tangent 

plt.plot(x, np.sin(x), '-g', label ='sin(x)')
plt.plot(x, np.cos(x), ':b', label = 'cos(x)')
plt.plot(x,np.tan(x), '-k', label = 'tan(x)')
plt.axis('equal')
plt.legend()
```

```python
#this is how you shape your graph and scatter it 

x = np.linspace(0,10,30)
y = np.sin(x)
plt.plot(x,y,'o', color = 'red');
```

```python
#another way to create legends the numpoints means 1 key in legends 

rng = np.random.RandomState(0)
for maker in ['o','-',',','x','+','v','<','>','s', 'd']:
  plt.plot(rng.rand(5),rng.rand(5), maker,
           label = "maker='{0}'".format(maker))
  plt.legend(numpoints = 1)
  plt.xlim(0,1,8)
```

```python

#alpha represent transparency this is another way you can do a 
rng  = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000*rng.rand(100)
plt.scatter(x,y,c =colors, s = sizes, alpha =0.3, cmap = 'viridis')
plt.colorbar(); #show color scale 
```
